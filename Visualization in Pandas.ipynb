# Visualization in Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\david.akor\Downloads\DAP\Ice Cream Ratings.csv")
df =df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('classic')
```

```python
df.plot()
```

```python
df.plot(kind = 'line')
```

```python
df.plot(kind = 'line', subplots=True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel='Daily Ratings',ylabel='Scores',figsize=(10,5))
```

```python
df.plot(kind ='bar',figsize=(10,5))
```

```python
df.plot(kind ='bar', stacked=True)
```

```python
df['Flavor Rating'].plot(kind ='bar', stacked=True)
```

```python
df.plot.barh( stacked=True,xlabel='Scores')
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=500, c='Brown')
```

```python
df.plot.hist()
```

```python
df.plot.hist(bins=10)
```

```python
df.boxplot()
```
